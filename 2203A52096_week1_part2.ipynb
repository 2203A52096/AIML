```python
#Part 2 â€“ Visualization of Data using matplotlib, pyplots Packages


import matplotlib.pyplot as plt
import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun']
ages = [25, 30, 35, 40]
heights = [145, 151, 165, 173]
weights = [45, 55, 65, 75]

data = {
    'Person': people,
    'Age': ages,
    'Height': heights,
    'Weight': weights
}

healthcare_dataset = pd.DataFrame(data)

print(healthcare_dataset)
```

```python
# Scatter plot of Height vs Weight
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(data=healthcare_dataset, x='Height', y='Weight', hue='Person', s=100)
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# Bar chart of Average Height and Weight
average_height = healthcare_dataset.groupby('Person')['Height'].mean()
average_weight = healthcare_dataset.groupby('Person')['Weight'].mean()

plt.figure(figsize=(10, 6))
average_height.plot(kind='bar', color='red', alpha=0.7, label='Average Height')
average_weight.plot(kind='bar', color='black', alpha=0.7, label='Average Weight')
plt.title('bar graph')
plt.xlabel('Person')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

```python
# Histogram of Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=healthcare_dataset, x='Age', bins=10, kde=True, color='green')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
